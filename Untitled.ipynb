```python
#import the libraries we will need


import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import datetime as dt
```

```python
#now read all the files that we have for this project


test_data = pd.read_csv('Test.csv',parse_dates=['TransactionStartTime','IssuedDateLoan'])
train_data = pd.read_csv('Train.csv',parse_dates=['TransactionStartTime','IssuedDateLoan'])
unlinked_mask = pd.read_csv('unlinked_masked_final.csv',parse_dates=['TransactionStartTime'])
```

```python
#get an overview of the data

test_data.head(20).transpose()
```

```python
# FILTER THE TRAIN DATA AND FILTER ONLY THE IMPORTANT COLUMNS ONLY


train_modified = train_data[['CustomerId','Value','Amount','ProductId','IsDefaulted','BatchId','InvestorId','SubscriptionId','ChannelId','TransactionStartTime','IssuedDateLoan','ThirdPartyId']]
```

```python
train_modified.describe()
```

```python
#FUNCTION TO CONVERT THE BATCH ID'S INTO A NUMBER; SMALL NUMBER


def label_converter(string):
    return (int(string.split('_')[1]))/100

def zero_ones(string):
    if string == 'nan':
        return(0)
    else:
        return(1)
```

```python
train_modified['BatchId'] = train_modified.BatchId.apply(label_converter)
```

```python
#MODIFIE THE VALUE AND AMOUNT COLUMNS SO THAT THEY BECOME SMALL NUMBERS


train_modified['Value'] = train_modified['Value']/5000
train_modified['Amount'] = train_modified['Amount']/5000
```

```python
train_modified['ProductId'] = train_modified['ProductId'].apply(label_converter)
#train_modified['InvestorId'] = train_modified['InvestorId'].apply(label_converter)
train_modified['SubscriptionId'] = train_modified['SubscriptionId'].apply(label_converter)
train_modified['ChannelId'] = train_modified['ChannelId'].apply(label_converter)
```

```python
train_modified.head(15)
```

```python
train_modified.info()
```

```python
train_modified.head()
```

```python
#DROP ALL THE ROWS WITH NO VALUES


train_modified = train_modified.dropna()
```

```python
train_modified['InvestorId'] = train_modified['InvestorId'].apply(label_converter)
train_modified['ThirdPartyConfirmed'] = train_modified['ThirdPartyId'].apply(zero_ones)
```

```python
def get_month(x):
    return (x.month)
    
def get_year(x):
    return (x.year)/100
    
def get_day(x):
    return (x.day)
```

```python
train_modified['Tx_Start_month'] = train_modified['TransactionStartTime'].apply(get_month)
train_modified['Tx_Start_year'] = train_modified['TransactionStartTime'].apply(get_year)
train_modified['Tx_Start_day'] = train_modified['TransactionStartTime'].apply(get_day)

train_modified['loan_Start_month'] = train_modified['IssuedDateLoan'].apply(get_month)
train_modified['loan_Start_year'] = train_modified['IssuedDateLoan'].apply(get_year)
train_modified['loan_Start_day'] = train_modified['IssuedDateLoan'].apply(get_day)

train_modified = train_modified.drop(['TransactionStartTime','IssuedDateLoan'], axis = 1)
train_modified.head()
```

```python
train_modified.info()
```

```python
#LETS MAKE TWO GROUPS OUT OF THE PROVIDED DATA; ONE FOR TRAINING OUR MODEL AND THE OTHER TO VALIDATE THE MODEL


train = train_modified.reset_index().iloc[0:1000]

validation = train_modified.reset_index().iloc[1001:]
```

```python
train.info()
```

```python
#SEPARATE THE DATA AND THE TARGET 


x = train[['Value','Amount','ProductId','BatchId','InvestorId','SubscriptionId','ChannelId','Tx_Start_month','Tx_Start_year','Tx_Start_day','loan_Start_month','loan_Start_year','loan_Start_day','ThirdPartyConfirmed']]


y = train[['IsDefaulted']]

x.head()
```

```python
#TRAIN THE MODEL


model = RandomForestClassifier(n_estimators = 100)
model = model.fit(x, y.values.ravel())
```

```python
#PREP THE VALIDATION DATA


x_validation = validation[['Value','Amount','ProductId','BatchId','InvestorId','SubscriptionId','ChannelId','Tx_Start_month','Tx_Start_year','Tx_Start_day','loan_Start_month','loan_Start_year','loan_Start_day','ThirdPartyConfirmed']]

validation_predicted = model.predict(x_validation)
```

```python
y_actual = validation[['IsDefaulted']]
```

```python
model.score(x_validation, y_actual.values.ravel())
```

```python
wrong = 0
right = 0
for i in range(0,len(y_actual)):
    if y_actual.values[i][0] == validation_predicted.tolist()[i]:
        right += 1
    else:
        wrong += 1
    
```

```python
len(y_actual)
```

```python
print(wrong,right)
```

```python
right/(wrong + right)
```
